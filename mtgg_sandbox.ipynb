```python
m21 = "data/raw/m21_all_cards.html"

with open(m21) as fp:
    soup = BeautifulSoup(fp)

m21_tags = soup.select('span[class="card_name"]')
mtgg_link = "https://www.mtggoldfish.com/"
m21_card_links = []
for tag in m21_tags:
    #print (tag.find('a').get("href"))
    tmp_link = tag.find('a').get("href")
    tmp_link = mtgg_link+tmp_link
    m21_card_links.append(tmp_link)

drop_words = ['sealed']
m21_card_links_test = [word for word in m21_card_links if not any(bad in word for bad in drop_words)]
```

```python
# https://www.mtggoldfish.com/price-download/paper/Unifying+Theory+%255BOD%255D
# https://www.mtggoldfish.com/price-download/paper/Pyrohemia+%255BPLC%255D
```

```python
pyro = "data/raw/Pyrohemia.html"

with open(pyro) as fp:
    hotsoup = BeautifulSoup(fp)
```

```python
data  = hotsoup.find_all("script")
data[7].text
```

```python
import re
import pandas as pd

table = data[7].string

p = re.compile('d [+]= (.*?);', re.MULTILINE | re.DOTALL)
m = p.match(table)

p = re.compile('(d [+=][^g]*;)', re.MULTILINE | re.DOTALL)


# print(p.findall(data[7].text))

prices = p.findall(table)
#prices = pd.DataFrame(prices)
prices[0]
```

```python
test = prices[0].replace("\\n", "").replace("\n", "").replace("d = ", "").replace("d += ", "").replace('"', '').split(';  ')
test
```

```python
df = pd.DataFrame(test)[0].str.replace(';','').str.split(',', expand=True)
df
```

```python
df.rename(columns=df.iloc[0], inplace = True)
df.drop(df.index[0], inplace = True)
df
```

```python
df.iloc[:, 1] = pd.to_numeric(df.iloc[:, 1],errors='coerce')
```
